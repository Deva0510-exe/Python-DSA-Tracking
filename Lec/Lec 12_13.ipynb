```python
## if else statement 
age = int(input("Enter your age: "))

if age>=18:
    print("you are eligible")
else:
    print("not eligible")
```

```python
## elif statement

age = int(input("Enter your age: "))

if age<=12:
    print("you are child")
elif age<=18:
    print("you are teenager")
else:
    print("Your Adult")
```

```python
## Nested Conditional Statement
x=int(input("Enter any Number"))

if x>=0:
    if x%2==0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Negative Number")
```
